### Data cleaning

#### Handling duplicates

```python
df.duplicated().sum()
```

```python
df.columns = df.columns.str.strip().str.title()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df[df['Director'].isnull()].head()
```

```python
df['Director']=df['Director'].fillna('Unknown')
```

```python
df['Cast']=df['Cast'].fillna('Unknown')
```

```python
df['Country']=df['Country'].fillna('Unknown')
```

```python
df['Date_Added'].isnull().sum()
```

```python
df=df.dropna(subset=['Date_Added'])
```

```python
df['Rating']=df['Rating'].fillna('Not Rated')
```

```python
df['Rating'].isnull().sum()
```

```python
df[df['Rating']=='Not Rated'].head()
```

```python
df['Duration'].isnull().sum()
```

```python
df[df['Duration'].isnull()].head()
```

```python
mask = df['Duration'].isnull()
df.loc[mask,'Duration']=df.loc[mask,'Rating']
```

```python
df[df['Rating']=='Not Rated'].head()
```

```python
df.loc[mask, 'Rating']='Not Rated'
```

```python
df[df['Rating']=='Not Rated'].head(10)
```

```python
string_cols=df.select_dtypes('str').columns
```

```python
df[string_cols]=df[string_cols].apply(lambda x:x.str.strip())
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Date_Added']
```

```python
df['Date_Added'] = pd.to_datetime(df['Date_Added'],format='mixed')
```

```python
df['Date_Added']
```

#### Data Cleaning Summary

- Cleaned column names by removing leading and trailing spaces and converting them to title case.
- Filled missing values in the 'Director', 'Cast', and 'Country' columns with 'Unknown'.
- Removed rows with missing values in the 'Date_Added' column.
- Corrected missing values in the 'Duration' column by transferring misplaced duration values from the 'Rating' column.
- Labelled missing values in the 'Rating' column as 'Not Rated'.
- Removed leading and trailing spaces from all object-type columns.
- Converted the 'Date_Added' column to datetime format.

### EDA(Exploratory Data Analysis)

```python
import matplotlib.pyplot as plt
```

```python
df['Type'].value_counts().plot(kind='bar')
plt.title("Distribution by Type")
plt.xlabel('Type of Content')
plt.ylabel('Number of Titles')
plt.savefig('Distribution by Type.png', bbox_inches='tight')
plt.show()
```

#### Significantly more Movies than TV Shows.

```python
df['Release_Year'].value_counts().sort_index().plot(figsize=(12,5))
plt.title("Title release by year")
plt.xlabel('Year when Title released')
plt.ylabel('Number of Titles')
plt.savefig('Number of Titles release by Year.png', bbox_inches='tight')
plt.show()
```

#### Number of title released increased drastically after 2010. 

```python
df[df['Country']!='Unknown']['Country'].value_counts().head(10).plot(kind='bar')
plt.title('Most contributing Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.savefig('Most contributing countries.png', bbox_inches='tight')
plt.show()
```

#### United States contributes the largest number of titles on Netflix.

```python
df['Rating'].value_counts().head(10).plot(kind='bar')
plt.title('Highest Rating categories')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.savefig('Highest rating categories.png', bbox_inches='tight')
plt.show()
```

#### TV-MA is the most common rating category.

```python
df['Listed_In'].value_counts().head(10).plot(kind='bar')
plt.title('Most common Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.savefig('Most Common Genres.png', bbox_inches='tight')
plt.show()
```

#### Dramas,International Movies is the Most common Genre.

```python
df['Year_Added']=df['Date_Added'].dt.year
df['Year_Added'].value_counts()
```

```python
df['Year_Added'].value_counts().sort_index().plot()
plt.title('Years when content added')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.savefig('Year with contents added.png', bbox_inches='tight')
plt.show()
```

#### Rapid growth in number of title after 2015.

```python
movies=df[df['Type']=='Movie'].copy()
```

```python
movies['Duration_minutes']=movies['Duration'].str.extract(r'(\d+)').astype(int)
```

```python
Avg_duration = round(movies['Duration_minutes'].mean(),2)
print(f'Average_minutes:{Avg_duration} minutes')
```

```python
movies['Duration_minutes'].hist()
plt.title('Average duration of Movies')
plt.xlabel('Duration in Minutes')
plt.ylabel('Number of Movies')
plt.savefig('Average Duration of Movies', bbox_inches='tight')
plt.show()
```

#### Average duration of movies is 99.56 minutes
